```python
import pandas as pd

df = pd.read_csv('/content/flights_sample_3m.csv')
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.columns.tolist()
```

```python
#Drop Unnecessary Columns
df = df.drop(columns=[
    'AIRLINE_DOT',
    'AIRLINE_CODE',
    'DOT_CODE',
    'DIVERTED',
])
```

```python
#Drop Rows with Nulls in Key Columns
df = df.dropna(subset=[
    'DEP_DELAY', 'ARR_DELAY', 'DEP_TIME', 'ARR_TIME',
    'CRS_DEP_TIME', 'CRS_ELAPSED_TIME', 'ORIGIN', 'DEST'
])
```

```python
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])
df['Month_Year'] = df['FL_DATE'].dt.to_period('M')
df.shape
```

```python
df.head()
```

```python
df.to_csv('cleaned_flights.csv', index=False)
```
